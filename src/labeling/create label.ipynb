## Create labeling
- 클래스 별로 데이터 인덱스가 다른 라벨링

```python
import random
import math
seed_num = 2    # random 시드 설정
random.seed(seed_num)

# 총 데이터 수
total_num1 = 81
total_num2 = 6480
total_num3 = 7980

# 사용할 데이터 수
using_num1 = 81
using_num2 = 1000
using_num3 = 1300

# (학습 데이터 수 : 테스트 데이터 수) 비율 설정
tr_ratio = 9   
ts_ratio = 10-tr_ratio

label_num = {"bsn":0, "chw":1, "kbj":2, "kjh":3, "ljh":4, "lse":5, "pjh":6, "psm":7, "shb":8, "sws":9}
label_name = {0:"bsn", 1:"chw", 2:"kbj", 3:"kjh", 4:"ljh", 5:"lse", 6:"pjh", 7:"psm", 8:"shb", 9:"sws"}
# name = "bsn"        # 이름 수정

tr_list = []
ts_list = []

## 총 데이터 수에서 사용할 데이터 수를 랜덤으로 뽑고, 정렬하고, 랜덤으로 섞는다.
## 섞인 데이터를 비율에 따라 학습 데이터, 테스트 데이터로 나눈다.

for j in range(10):
    name = label_name[j]

    # 1 Syllable data labeling #
    index1 = sorted(random.sample(range(total_num1), using_num1))
    random.shuffle(index1)  # 0~80 중 81개 랜덤 선택

    n1 = math.ceil(using_num1 * tr_ratio/(tr_ratio+ts_ratio))   # Training data 개수
    n2 = math.floor(using_num1 * ts_ratio/(tr_ratio+ts_ratio))  # Test data 개수

    for i in range(using_num1):
        if i < n1:
            string = "./dataset/" + name + "/1/" + name + "_" + str(index1[i]) + ".jpg"
            string = string + " " + str(label_num[name])
            tr_list.append(string)
        else:
            string = "./dataset/" + name + "/1/" + name + "_" + str(index1[i]) + ".jpg"
            string = string + " " + str(label_num[name])
            ts_list.append(string)


    # 2 Syllable data labeling #
    index2 = sorted(random.sample(range(total_num2), using_num2))
    random.shuffle(index2)

    n1 = math.ceil(using_num2 * tr_ratio/(tr_ratio+ts_ratio))   # Training data 개수
    n2 = math.floor(using_num2 * ts_ratio/(tr_ratio+ts_ratio))  # Test data 개수

    for i in range(using_num2):
        if i < n1:
            string = "./dataset/" + name + "/2/" + name + "2_" + str(index2[i]) + ".jpg"
            string = string + " " + str(label_num[name])
            tr_list.append(string)
        else:
            string = "./dataset/" + name + "/2/" + name + "2_" + str(index2[i]) + ".jpg"
            string = string + " " + str(label_num[name])
            ts_list.append(string)

    # 3 Syllable data labeling #
    index3 = sorted(random.sample(range(total_num3), using_num3))
    random.shuffle(index3)

    n1 = math.ceil(using_num3 * tr_ratio/(tr_ratio+ts_ratio))   # Training data 개수
    n2 = math.floor(using_num3 * ts_ratio/(tr_ratio+ts_ratio))  # Test data 개수

    for i in range(using_num3):
        if i < n1:
            string = "./dataset/" + name + "/3/" + name + "3_" + str(index3[i]) + ".jpg"
            string = string + " " + str(label_num[name])
            tr_list.append(string)
        else:
            string = "./dataset/" + name + "/3/" + name + "3_" + str(index3[i]) + ".jpg"
            string = string + " " + str(label_num[name])
            ts_list.append(string)
        
f1 = open("list_tr1",'w')
for s in tr_list:
    f1.write("%s\n" %s)

f2 = open("list_ts1",'w')
for s in ts_list:
    f2.write("%s\n" %s)

f1.close()
f2.close()
```
