## 2글자 데이터 만들기

```python
import cv2
import numpy as np
import itertools

def img_concat(img1, img2):
    h1, w1, c1 = img1.shape
    h2, w2, c2 = img2.shape
    
    if h1 > h2:
        scale = h1/h2
        x = int(w2 * scale)  # 너비
        y = h1          # 높이
        img22 = cv2.resize(img2, dsize=(x, y), interpolation=cv2.INTER_AREA)
        result = cv2.hconcat([img1, img22])
    else:
        scale = h2/h1
        x = int(w1 * scale)  # 너비
        y = h2          # 높이
        img11 = cv2.resize(img1, dsize=(x, y), interpolation=cv2.INTER_AREA)
        result = cv2.hconcat([img11, img2])
        
    return result

def img_enlarge(img):
    h, w, c = img.shape

    if w >= h: 
        length = int((w-h)/2)
        tmp_img = np.ones((length, w, 3),dtype=np.uint8) * 255
        result = cv2.vconcat([tmp_img, img, tmp_img])
    else:
        length = int((h-w)/2)
        tmp_img = np.ones((h, length, 3),dtype=np.uint8) * 255
        result = cv2.hconcat([tmp_img, img, tmp_img])

    return result

name = "pjh"        # Class name 수정

# 81개 이미지를 리스트에 저장 #
img_arr = []
for i in range(81):
    path = "./dataset/" + name + "/1x/" + name + "_" + str(i) + "x.jpg"
    img = cv2.imread(path)
    img_arr.append(img)

nPr = itertools.permutations(img_arr, 2)    # 6480개의 이미지 순열 생성 (81P2)

j = 0
for case in nPr:
    img1 = case[0]
    img2 = case[1]
    
    img0 = img_concat(img1, img2)
    img00 = img_enlarge(img0)
    
    img_re = cv2.resize(img00, dsize=(64, 64), interpolation=cv2.INTER_AREA)
    
    path = "./dataset/" + name + "/2/" + name + "2_" + str(j) + ".jpg"
    j = j + 1
    cv2.imwrite(path, img_re)
    
```

## 3글자 데이터 만들기

```python
import cv2
import numpy as np
import itertools

def img_concat(img1, img2):
    h1, w1, c1 = img1.shape
    h2, w2, c2 = img2.shape
    
    if h1 > h2:
        scale = h1/h2
        x = int(w2 * scale)  # 너비
        y = h1          # 높이
        img22 = cv2.resize(img2, dsize=(x, y), interpolation=cv2.INTER_AREA)
        result = cv2.hconcat([img1, img22])
    else:
        scale = h2/h1
        x = int(w1 * scale)  # 너비
        y = h2          # 높이
        img11 = cv2.resize(img1, dsize=(x, y), interpolation=cv2.INTER_AREA)
        result = cv2.hconcat([img11, img2])
        
    return result

def img_enlarge(img):
    h, w, c = img.shape

    if w >= h: 
        length = int((w-h)/2)
        tmp_img = np.ones((length, w, 3),dtype=np.uint8) * 255
        result = cv2.vconcat([tmp_img, img, tmp_img])
    else:
        length = int((h-w)/2)
        tmp_img = np.ones((h, length, 3),dtype=np.uint8) * 255
        result = cv2.hconcat([tmp_img, img, tmp_img])

    return result

name = "sws"        # Class name 수정

# 81개 이미지를 리스트에 저장 #
img_arr = []
for i in range(0,21):       # 6480개 생성   0~20 : 21P3 = 7980개
    path = "./dataset/" + name + "/1x/" + name + "_" + str(i) + "x.jpg"
    img = cv2.imread(path)
    img_arr.append(img)

nPr = itertools.permutations(img_arr, 3)    # nP3

j = 0
for case in nPr:
    img1 = case[0]
    img2 = case[1]
    img3 = case[2]
    
    img0 = img_concat(img1, img2)
    img00 = img_concat(img0, img3)
    img000 = img_enlarge(img00)
    
    img_re = cv2.resize(img000, dsize=(64, 64), interpolation=cv2.INTER_AREA)
    
    path = "./dataset/" + name + "/3/" + name + "3_" + str(j) + ".jpg"
    j = j + 1
    cv2.imwrite(path, img_re)
```

```python
import cv2
import numpy as np

name = "kjh"

for i in range(7980):
    path = "./dataset/"+ name +"/3/tmp/" + name + "3_" + str(i) + ".jpg"
    img = cv2.imread(path)

    # Unsharped mask 적용 #
    t_rimg = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
    Y, Cr, Cb = cv2.split(t_rimg)
    blur = cv2.GaussianBlur(Y, (0,0), 2)
    filtered_Y = np.clip(2.0*Y - blur, 0, 255).astype(np.uint8)
    img_unsharped = cv2.cvtColor(cv2.merge((filtered_Y, Cr, Cb)), cv2.COLOR_YCrCb2BGR)

    path2 = "./dataset/"+ name +"/3/" + name + "3_" + str(i) + ".jpg"
    cv2.imwrite(path2, img_unsharped)

# cv2.imshow("img", img)
# cv2.imshow("img_unsharped", img_unsharped)


# key = cv2.waitKey(0)
# cv2.destroyAllWindows()
```
